# Daisy feature testing

```python
import cv2
import os
import skimage.feature 
from FeatureTransformer import FeatureTransformer 
```

```python
transformer = FeatureTransformer()
daisy_params = {
    'step': 32,
    'radius': 8,
    'rings': 2,
    'histograms': 8,
    'orientations': 8,
    'visualize': False 
}
```

# 2. Feature extraction

```python
x_train={}
y_train={}
```

```python
def process_images_in_folder(class_folder_path,class_folder_name,x_dict,y_dict):
    if os.path.isdir(class_folder_path):
        # Iterate through all the image files in the class folder
        for filename in os.listdir(class_folder_path):
            if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif')):
                # Read the image using OpenCV
                image_path = os.path.join(class_folder_path, filename)
                image = cv2.imread(image_path)
                image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                name= str(class_folder_name)+str(filename)
                print(name)
                
                if image is not None:
                    # Extract daisy feature
                    daisy = transformer.transform(image=image,method=skimage.feature.daisy,params=daisy_params)
                    x_dict[name]= [daisy]
                    y_dict[name]= 0  #nombre de la classe
                    
```

```python
train_folder_path = 'data/train'
test_folder_path = 'data/test'

i = 0
# Extract features from training data
for class_folder in os.listdir(train_folder_path):
    print(class_folder)
    class_folder_path = os.path.join(train_folder_path, class_folder)
    process_images_in_folder(class_folder_path,class_folder,x_train,y_train)
    i=i+1
    if(i==100):
        break
    

# Extract features from testing data
#for class_folder in os.listdir(test_folder_path):
 #   print(class_folder)
  #  class_folder_path = os.path.join(test_folder_path, class_folder)
   # process_images_in_folder(class_folder_path,class_folder,x_test,y_test)
```
